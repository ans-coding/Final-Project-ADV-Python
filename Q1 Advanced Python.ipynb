```python
def stability(f):
    if f>0.03:
        return("unstable")
    elif f<=0.03:
        return("stable")
```

```python
def bandgaptype(b):
    if b==0:
        return("metal")
    elif b>0 and b<=1:
        return("small band")
    elif b>1:
        return("large band")
```

```python
def sbgt(x,y):
    if x>0.03:
        a = "unstable"
    elif x<=0.03:
        a ="stable"
    if y==0:
        b = "metal"
    elif y>0 and b<=1:
        b = "small band"
    elif y>1:
        b = "large band"
    return(a,b)
def sbgt_row(r):
    return sbgt(r['formation_energy_per_atom'], r['band_gap'])
```

```python
df['stability'] = df['formation_energy_per_atom'].apply(stability)
```

```python
df['band_gap_type'] = df['band_gap'].apply(bandgaptype)
```

```python
df['joint'] = df.apply(lambda r: (stability(r['formation_energy_per_atom']),bandgaptype(r['band_gap'])), axis=1)
```
